## 4) Check and remove duplicates


```python
df.drop_duplicates(inplace=True)
df.shape
```

## 5) Rename columns ("Title" > "Movie_Title")

```python
df.rename(columns={"Title": "Movie_Title"}, inplace=True)
df.head()
```

## 6) Get unique genres


```python
genres = df['Genre'].str.split(',').explode()
unique_genres = genres.unique()
unique_genres
```

## 7) Extract number of genres per movie


```python
df['Num_Genres'] = df['Genre'].str.split(',').apply(len)
df.head()
```

## 8) Handle missing values in 'Revenue (Millions)' and 'Metascore


```python
df['Revenue (Millions)'] = df['Revenue (Millions)'].fillna(df['Revenue (Millions)'].mean())
df['Metascore'] = df['Metascore'].fillna(df['Metascore'].mean())

print(df[['Revenue (Millions)','Metascore']].head())
print(df[['Revenue (Millions)','Metascore']].isnull().sum())
```

## 9) Top 5 movies by Rating


```python
top_5_movies_by_rating = df.sort_values(by='Rating', ascending=False).head(5)
top_5_movies_by_rating
```

## 10) Movies with highest/lowest revenue


```python
highest_revenue_movies = df.sort_values(by='Revenue (Millions)', ascending=False).head(5)
highest_revenue_movies
lowest_revenue_movies = df.sort_values(by='Revenue (Millions)', ascending=True).head(5)
lowest_revenue_movies
```

## 11) Avg. rating by Director


```python
avg_rating_by_director = df.groupby('Director')['Rating'].mean().reset_index()
avg_rating_by_director.columns = ['Director', 'Average_Rating']
avg_rating_by_director
```

## 12) Movies per year


```python
movies_per_year = df['Year'].value_counts().sort_index()
movies_per_year
```

## 13) Movies with rating > 8 & revenue > 100M

```python
movies_filtered = df[(df['Rating'] > 8) & (df['Revenue (Millions)'] > 100)]
movies_filtered
```

## 14) Group by Director: avg Rating, Revenue


```python
group_by_director = df.groupby('Director')[['Rating', 'Revenue (Millions)']].mean().reset_index()
group_by_director.columns = ['Director', 'Average_Rating', 'Average_Revenue']
group_by_director
```

## 15) Group by Year: count, avg Rating


```python
group_by_year = df.groupby('Year').agg(
    Count=('Movie_Title', 'size'),
    Average_Rating=('Rating', 'mean')
).reset_index()
group_by_year
```

16) Create a new column Profitability:

a) Revenue divided by Duration

b) Find top 10 most profitable movies


```python
df['Profitability'] = df['Revenue (Millions)'] / df['Runtime (Minutes)']
top_10_profitable_movies = df.sort_values(by='Profitability', ascending=False).head(10)
top_10_profitable_movies[['Movie_Title', 'Revenue (Millions)', 'Runtime (Minutes)', 'Profitability']]
```

## 17) Create a column that categorizes movies by rating:
a) Rating < 6: Poor

b) 6  Rating < 7.5: Average

```python
df['Rating_Category'] = df['Rating'].apply(lambda x: 'Poor' if x < 6 else ('Average' if 6 <= x < 7.5 else 'Good'))
df.head()
```
