##7. Replace missing age values with the median age.

##8. Create a column age_group with values:
a. 'Child' for age < 18

b. 'Adult' for age 1860

c. 'Senior' for age > 60

d. Use: apply() or np.where() or custom function

```python
df['age'] = df['age'].fillna(df['age'].median())
def age_group_classifier(age):
    if age < 18:
        return 'Child'
    elif age <= 60:
        return 'Adult'
    else:
        return 'Senior'

df['age_group'] = df['age'].apply(age_group_classifier)
df.head()
```
