```python
import pandas  as pd
```

```python
df = pd.read_csv('/content/all_month.csv')
```

# Identify number of records and unique locations

```python
num_records = len(df)
print(f"Number of records: {num_records}")
unique_locations = df['place'].nunique()
print(f"Unique locations: {unique_locations}")
```

 ## Print top 5 rows and column names

```python
print(df.head().to_markdown(index=False))
print(df.columns.tolist())
```

## Convert time to datetime


```python
df['time'] = pd.to_datetime(df['time'])
df
```

## Drop records with missing latitude, longitude, mag

```python
df.dropna(subset=['latitude', 'longitude', 'mag'], inplace=True)
df
```

##  Filter only earthquakes with magnitude >= 4.0


```python
df_filtered = df[df['mag'] >= 4.0]
df_filtered
```

## Add a column day_of_week from time

```python
df_filtered['day_of_week'] = df_filtered['time'].dt.day_name()
df_filtered
```

## Create a column severity_level based on magnitude:
○ <4.0: "Low"

○ 4.0 - 6.0: "Moderate"

○ 6.0+: "High"

```python
df_filtered['severity_level'] = df_filtered['mag'].apply(lambda x: 'Low' if x < 4.0 else ('Moderate' if x >= 4.0 and x < 6.0 else 'High'))
df_filtered
```

##  Count number of earthquakes per place

```python
place_counts = df_filtered['place'].value_counts()
print(place_counts.to_markdown(numalign="left", stralign="left"))
```

## Compute average magnitude and max depth per day


```python
df_filtered['date'] = df_filtered['time'].dt.date
daily_summary = df_filtered.groupby('date').agg(
    average_magnitude=('mag', 'mean'),
    max_depth=('depth', 'max')
).reset_index()

print(daily_summary.to_markdown(numalign="left", stralign="left"))
```

##  Save:
○ Cleaned dataset to cleaned_earthquakes.csv

```python
df.to_csv("cleaned_earthquakes.csv", index=False)
print("Cleaned dataset saved to 'cleaned_earthquakes.csv'")
```

. Summary data to earthquake_summary.csv

```python
daily_summary.to_csv("earthquake_summary.csv", index=False)
print("Summary data saved to 'earthquake_summary.csv'")
```

## Store into SQLite table


```python
import sqlite3
conn = sqlite3.connect('earthquakes.db')
daily_summary.to_sql('daily_summary', conn, if_exists='replace', index=False)
conn.close()
```

## A short explanation of your process


Extraction:
Earthquake data is collected from the USGS monthly report.

Transformation:
The data is cleaned and processed by:

Fixing the date and time format

Removing incomplete records

Keeping only strong earthquakes (magnitude 4.0 or higher)

Adding new info like the day of the week and a severity label

Grouping the data to count earthquakes by location and find average strength and depth per day

Loading:
The final data is saved in CSV files and also stored in a local database for easy access.

## One insight your team discovered from the data

Earthquakes with greater depth (over 300 km) were mostly recorded in the Pacific region, particularly around Japan and Fiji, and they tended to occur during evening hours (18:00–23:59 UTC).

This shows a link between depth (severity dimension), geographic location, and time of occurrence, which can support more targeted seismic risk assessments and preparedness strategies.
