```python
import pandas as pd
```

```python
#to import excel sheets from sheet1 to sheet2

df = pd.read_excel('/content/employee.xlsx',sheet_name="Sheet2")
df.head()
```

```python
df = pd.read_excel('/content/employee.xlsx',sheet_name="Sheet1")
df.head()
```

```python
# combine sheet1 an sheet2

df1 = pd.read_excel('/content/employee.xlsx',sheet_name="Sheet1", skiprows=1)
df2 = pd.read_excel('/content/employee.xlsx',sheet_name="Sheet2")
result = pd.concat([df1,df2],ignore_index=True)
display(result)
```

```python
# to combine excel sheets and in another file

result.to_excel('result.xlsx')
```
