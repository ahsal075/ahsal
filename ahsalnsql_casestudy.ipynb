Step 1: Load SpaceX Launch Data from API

● Use https://api.spacexdata.com/v4/launches

● Extract relevant columns: name, date_utc, success, details, rocket

● Convert date_utc to datetime and extract the year

```python
import pandas as pd
```

```python
pip install requests
```

```python
import requests
url1 = 'https://api.spacexdata.com/v4/launches'
url2 = 'https://api.spacexdata.com/v4/rockets'

response1 = requests.get(url1)
response2 = requests.get(url2)

print("Response for launches:")
print(response1)
print("\nResponse for rockets:")
print(response2)
```

```python
launch = response1.json()
launch = pd.DataFrame(launch)
launch.head()
```

```python
dflaunch = launch[['name', 'date_utc', 'success', 'details', 'rocket']]
dflaunch
```

```python
dflaunch ['date_utc']=pd.to_datetime(dflaunch['date_utc'])
dflaunch
```

```python
dflaunch ['year']=dflaunch['date_utc'].dt.year
dflaunch
```

Step 2: Load Rocket Metadata

● Use https://api.spacexdata.com/v4/rockets

● Extract id, name, type, active, and stages


```python
url2 = 'https://api.spacexdata.com/v4/rockets'
response2 = requests.get(url2)
print(response2)
```

```python
rocket = response2.json()
rocket = pd.DataFrame(rocket)
rocket.head()
```

```python
dfrocket = rocket[['id', 'name', 'type', 'active', 'stages']]
dfrocket
```

```python
print(rocket.columns)
```

Step 3: Merge Launch and Rocket Data]

● Join the two DataFrames on rocket ID using


```python
mergeddf = pd.merge(dflaunch,dfrocket,left_on='rocket',right_on='id')
mergeddf
```

```python
mergeddf=mergeddf.drop(['rocket'],axis=1)
mergeddf
```

 Add Simulated Country Information

● Add a new column country and randomly assign one of these values:

['USA', 'Russia', 'India', 'China', 'France']


```python
import random
```

```python
countries = ['USA', 'Russia', 'India', 'China', 'France']
mergeddf['country'] = [random.choice(countries) for _ in range(len(mergeddf))]
mergeddf
```

Step 5: Store Merged Data in SQLite3

● Use sqlite3 to create a connection and save the merged DataFrame as a table named
launches

● Table should contain all merged columns including country

```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect('launches.db')
mergeddf.to_sql('launches',conn,if_exists='replace',index=False)
```

```python
pd.read_sql_query('SELECT * FROM launches LIMIT 5;',conn)
```

Step 6: Run SQL Queries on the Data to analyze

1. Launches by Country

2. Which year had the highest number of launches?

3. Top 5 Missions by Launch Count


```python
query1 = """
SELECT country, COUNT(*) AS launch_count
FROM launches
GROUP BY country
ORDER BY launch_count DESC;"""

pd.read_sql_query(query1,conn)
```

```python
query2 = """
SELECT year, COUNT(*) AS launch_count
FROM launches
GROUP BY year
ORDER BY launch_count DESC LIMIT 1;"""

pd.read_sql_query(query2,conn)
```

```python
query3 = """
SELECT name_y AS mission_name, COUNT(*) AS mission_count
FROM launches
GROUP BY name_y
ORDER BY COUNT(*) DESC LIMIT 5;"""

pd.read_sql_query(query3,conn)
```
